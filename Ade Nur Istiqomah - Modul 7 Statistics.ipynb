## No. 1 Who is the youngest and oldest player in the NBA in 2017 for each team (Tm)?
Columns to keep: Player, Age, Tm.

```python
dataset_1 = dataset[['Player', 'Age', 'Tm']]
dataset_1.head()
```

```python
#How many rows are there?
print(len(dataset_1))
```

```python
#Any NaN value?
dataset_1.count()
```

```python
#Find the oldest and youngest age for each team
age_stat = dataset_1.groupby('Tm').agg({'Age': ['min', 'max']})
age_stat
```

```python
#Check whether each age value corresponds to only one player
count_player_age = dataset_1.groupby(['Tm', 'Age']).size()
count_player_age
```

Now we know that at least one team (ATL) has two youngest player with the same age.

```python
player_age_min = dataset_1.groupby('Tm').min()
player_age_min
```

```python
player_age_max = dataset_1.groupby('Tm').max()
player_age_max
```

## No. 3 Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)?
Columns to keep: Team, TRB%, AST%, STL%, BLK%

```python
dataset_3 = dataset[['Tm', 'TRB%', 'AST%', 'STL%', 'BLK%']]
dataset_3.head()
```

```python
dataset_3_ave_stat = dataset_3.groupby('Tm').mean()
dataset_3_ave_stat
```

```python
dataset_3_ave_stat.sort_values(by=['TRB%', 'AST%', 'STL%', 'BLK%'], ascending = False)
```

```python
dataset_3_ave_stat.idxmax()
```

## No. 4 Who is the best player in your opinion based on his record stats? note: you can refer to variables point (PTS), assists, rebounds, or anything else. A combination of several variables would be nice.

Quick research on how does people mostly determine NBA's best player by the number, we decided to take into account these parameters:
1. Points
2. Rebounds
3. Assists
4. Blocks
5. G
6. Steals

```python
best_player = dataset[['Player', 'Tm', 'PTS', 'TRB', 'AST', 'STL', 'BLK', 'G']]
best_player
```

Simply take the average based on those parameter.

```python
best_player['mean'] = best_player.mean(axis=1)
best_player = best_player.sort_values('mean', ascending = False)
```

```python
print("The best player is: " + str(best_player.iloc[0]))
```
